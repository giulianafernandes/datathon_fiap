## Limpeza dos dados

```python
dados.columns = map(str.lower, dados.columns) #transformando o título das colunas em lowercase
dados = dados.rename(columns={"nome": "id_aluno"}) # transformando o nome do aluno em um id
dados['id_aluno'] = dados['id_aluno'].str.replace("ALUNO-", "pm-").astype(str)
first_column = dados.pop('id_aluno') # colocando o id na primeira posição do dataframe
dados.insert(0, 'id_aluno', first_column)
dados.head(3)
```

```python
dados.instituicao_ensino_aluno_2020.unique()
```

```python
dados.idade_aluno_2020.unique() 
```

```python
# existe uma idade inconsistente nessa coluna, vamos localizar e tentar fazer sentido dela
idade_inconsistente = dados['idade_aluno_2020'] == 'D108'
dados[idade_inconsistente]
```

```python
# como essa idade não faz sentido e o aluno aparentemente parou de frequentar 
# a passos mágicos em 2021, decidi remover essa linha
dados = dados.drop(1258)
```

```python
dados.idade_aluno_2020.unique()
```

```python
dados['idade_aluno_2020'].isna().sum()
```

```python
# existem 621 alunos com idade faltante no dataset
# vamos fazer uma transformação nos dados para encontrar o ano em que esses alunos nasceram
dados['idade_aluno_2020'] = dados['idade_aluno_2020'].fillna(2020) # preenchendo os dados nulos com 2020, para que na próxima transformação eles virem 0
dados['idade_aluno_2020'] = dados['idade_aluno_2020'].astype(int) # transformando o tipo da coluna
dados['ano_nascimento'] = 2020 - dados['idade_aluno_2020'] # criando uma nova coluna com o ano de nascimento do aluno
dados['ano_nascimento'].head(10)
```
